```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

2016 Data Sheet

```python
#Top 5 countries visited Sri Lanka in 2016

data16 = pd.read_csv('2016-P21.csv', index_col=None)
y16 = data16.iloc[:5, :2]

#Ploting the graph
plot = y16.plot.pie(y='TOTAL', figsize=(6, 6),)
plt.show()


y16
```

```python
#Top 5 countries visited SL in 2016
#Plotting Bar Chart

fig = plt.figure(figsize=(5,3))
ax = fig.add_axes([0,0,1,1])
COR = y16['Country of Residence']
pop = y16['TOTAL']
ax.bar(COR,pop)
```

Slicing only he details of UK from the dtatsheet

```python
#Slicing dtat seperately and print

uk_df16 = data16.iloc[2,2:]

#Ploting line chart
lines = uk_df16.plot.line()

uk_df16
```
